```julia
using Plots
```

```julia
#Plot for different values of α and x.
x = range(0.1, 5, length=50)
alphas = [0.5772, 1.6180, 2.7183, 3.1416]

plt = plot(title="Error E(x, α) for Various α", xlabel="x", ylabel="E(x, α)", legend=:topleft)

for alpha in alphas
    n = floor(Int, alpha)
    beta = alpha - n
    error = x .^ alpha .- ((x .^ (n + 1)) ./ ((1 - beta) .* x .+ beta))
    plot!(plt, x, error, label="α = $alpha")
end

display(plt)
savefig(plt, "latex_project/figures/error_plot.pdf")

```

```julia
#Plot error for fixed x and different values of α
x_val = 2.0
alphas = range(-1.9, 5, length=500)
errors = Float64[]

for alpha in alphas
    n = floor(Int, alpha)
    beta = alpha - n
    approx = (x_val^(n + 1)) / ((1 - beta) * x_val + beta)
    exact = x_val^alpha
    push!(errors, exact - approx)
end

plot_fixed_x = plot(alphas, errors,
    color="#ffb30e",
    xlabel="α",
    ylabel="E(x, α)",
    title="Error E(x, α) for fixed x = $x_val",
    legend=false,
    lw=2,
    grid=true)

hline!(plot_fixed_x, [0], color=:gray, linestyle=:dash, lw=1)

display(plot_fixed_x)
savefig(plot_fixed_x, "latex_project/figures/error_plot_fixed_x.pdf")
```
